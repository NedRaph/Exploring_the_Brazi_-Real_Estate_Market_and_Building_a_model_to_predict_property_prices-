## 1). Import necessary libraries

```python
import pandas as pd
import numpy as np
```

## 2). Loading the datasets

```python
df1=pd.read_csv("/workspaces/Exploring_the_Brazi_-Real_Estate_Market_and_Building_a_model_to_predict_property_prices-/data/brasil-real-estate-1.csv")
df2=pd.read_csv("/workspaces/Exploring_the_Brazi_-Real_Estate_Market_and_Building_a_model_to_predict_property_prices-/data/brasil-real-estate-2.csv")
```

## 3). Cleaning DataSet 1

```python
# Inspection
df1.info()
(df1.isna().sum()/len(df1))*100
duplicates_1=df1.duplicated().sum()
print(f"Total duplicated entries are {duplicates_1}")
df1.head()
```

```python
def wrangle_1(df):
    # 1. Drop nulls
    df=df.dropna()

    # 2. Clean "price_usd" column
    df["price_usd"]=(df["price_usd"]
                  .str.replace("$","", regex=False)
                  .str.replace(",", "", regex=False)
                  .astype(float)
                 )
    # 3. Clean "lat-lon"column
    df[["lat", "lon"]]=(df["lat-lon"]
                    .str.split(",", expand=True)
                    .astype(float)
                    )
    # 4. Clean "place_with_parent_name" column
    df["state"]=(df["place_with_parent_names"]
                .str.split("|", expand=True)[2]
                )

    # 5.Drop unnecessary columns
    df.drop(columns=["Unnamed: 0", "place_with_parent_names", "lat-lon"], inplace=True)
   
    return df
```

```python
df1=wrangle_1(df1)
df1.info()
df1.head()
```

## 4). Cleaning DataSet 2

```python
df2.info()
duplicate_2=df2.duplicated().sum()
print(f"The total number of duplicated entries is {duplicate_2}")
df2.head()
```

```python
# Drop nulls
def wrangle_2(df):
    # 1. Drop nulls
    df.dropna(inplace=True)

    # 2. Clean price column
    df["price_usd"]=df["price_brl"]*0.2

    # 3. Modify "area_m2" column
    df2["area_m2"]=df2.area_m2.astype(int)

    # 4. Drop unnecessary columns
    df2.drop(columns=["Unnamed: 0", "price_brl"], inplace=True)
    

    return df

```

```python
df2=wrangle_2(df2)
df2.info()
df2.head()
```

## 5). Concatenate the cleaned dataframes into one

```python
df=pd.concat([df2, df1], ignore_index=True)
df.info()
df.head()
```

## 6). Save the cleaned csv

```python
df.to_csv("brasil-real-estate-clean.csv", index=False)
```
